# Visualizing the Data for Better Understanding 

```python
df_filtered = df[df['State Name'] != 'Grand Total']

# Plotting the pie chart
plt.figure(figsize=(14, 14))
plt.pie(df_filtered['Electric Vehicle Count'], labels=None, autopct='%0.2f', startangle=90)
plt.legend(df_filtered['State Name'], title="States", loc="center left", bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
colors = sns.color_palette('viridis', len(df_filtered))

# Plotting the bar chart with colorful bars
plt.figure(figsize=(12, 6))
bars = plt.bar(df_filtered['State Name'], df_filtered['Electric Vehicle Count'], color=colors)
plt.xlabel('State')
plt.ylabel('Electric Vehicle Count')
plt.title('Electric Vehicle Count in Each State')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

# Adding a legend for better visualization
plt.legend(bars, df_filtered['State Name'], title="States", bbox_to_anchor=(1, 1), loc='upper left')

plt.show()
```

# visualization to compare cars vs range each car and there respective distance covered in each single charge

```python
# Sample data
cars = ["Tata Nexon EV", "Tata Tigor EV", "Tata Nexon EV Max", "MG ZS EV", "Hyundai Kona Electric", 
        "Jaguar I-Pace", "Audi E-Tron GT", "BYD E6", "Mercedes-Benz EQC", "BMW iX", 
        "Porsche Taycan", "Audi E-Tron"]

ranges = ["312 Km/Full Charge", "306 Km/Full Charge", "437 Km/Full Charge", "419 Km/Full Charge", 
          "452 Km/Full Charge", "470 Km/Full Charge", "388 Km/Full Charge", "415 Km/Full Charge", 
          "471 Km/Full Charge", "425 Km/Full Charge", "na", "400 Km/Full Charge"]

# Convert the range values to numeric for sorting
numeric_ranges = [int(range_val.split(" ")[0]) if range_val != "na" else 0 for range_val in ranges]

# Sort data based on range
sorted_data = sorted(zip(cars, numeric_ranges), key=lambda x: x[1], reverse=True)

cars, numeric_ranges = zip(*sorted_data)

# Plotting the bar chart
plt.figure(figsize=(12, 8))
plt.barh(cars, numeric_ranges, color='skyblue')
plt.xlabel('Range (Km/Full Charge)')
plt.title('Car vs Range')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

# visualization to show cars and there respective price

```python
import matplotlib.pyplot as plt
import re

# Sample data
cars = ["Tata Nexon EV", "Tata Tigor EV", "Tata Nexon EV Max", "MG ZS EV", "Hyundai Kona Electric", 
        "Jaguar I-Pace", "Audi E-Tron GT", "BYD E6", "Mercedes-Benz EQC", "BMW iX", 
        "Porsche Taycan", "Audi E-Tron"]

price_ranges = ["₹ 13.99 - 17.4 L", "₹ 12.49 - 13.64 L", "₹ 17.74 - 19.24 L", "₹ 21.99 - 25.88 L", 
                "₹ 23.79 - 23.98 L", "₹ 1.06 - 1.12 Cr", "₹ 1.8 Cr", "₹ 29.15 L", 
                "₹ 1 Cr", "₹ 1.16 Cr", "₹ 1.5 Cr", "₹ 1.01 - 1.19 Cr"]

# Convert the price range values to numeric for sorting
numeric_price_ranges = []

for price_range in price_ranges:
    # Extract numeric values from the price range using regular expression
    matches = re.findall(r'\d+\.\d+|\d+', price_range)
    if matches:
        numeric_price_ranges.append(float(matches[-1]))
    else:
        numeric_price_ranges.append(0.0)

# Sort data based on price range
sorted_data = sorted(zip(cars, numeric_price_ranges), key=lambda x: x[1], reverse=True)

cars, numeric_price_ranges = zip(*sorted_data)

# Plotting the horizontal bar chart
plt.figure(figsize=(12, 8))
plt.barh(cars, numeric_price_ranges, color='coral')
plt.xlabel('Price Range (INR)')
plt.title('Car vs Price Range')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```
